## Conditions (OR)

```python
qs = Shop.objects.all()
qs.filter(pk__gte=5, pk__lt=10) # AND
```

```python
from django.db.models import Q
```

Q(pk__gte=5)

```python
Q(pk__lt=10)
```

```python
Q(pk__gte=5) & Q(pk__lt=10)
```

```python
Q(pk__gte=5) | Q(pk__lt=10)
```

```python
qs = Shop.objects.all()
qs.filter(Q(pk__gte=5) | Q(pk__lt=10))
```

```python
conditions = Q(pk__gte=5)
conditions = conditions | Q(pk__lt=10)
```

```python
conditions = Q(pk__gte=5)
conditions |= Q(pk__lt=10)
```

## 지정 범위의 Rows

```python
qs = Shop.objects.all()
qs = qs.filter(pk__lte=10)
qs.update(description="not used description")
```

```python
qs = Shop.objects.all()
qs = qs.filter(pk__lte=10)
for shop in qs:
    shop.description = "not used description"
    shop.save()
```
